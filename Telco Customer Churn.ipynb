```python
import pandas as pd
import numpy as np
import seaborn as sns
data = pd.read_csv("customer_data.csv")
df = pd.DataFrame(data)
df
```

```python
df.drop('customerID',1, inplace=True)
df.replace(' ',0,inplace=True)
df.drop(0, 0, inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
obj_types = df.select_dtypes(include=['object']).copy()
obj_types.head()
```

```python
for column in obj_types:
    print(obj_types[column].value_counts(),'\n')    
```

```python
for column in obj_types:
    obj_types[column] = obj_types[column].astype('category')
    obj_types[column] = obj_types[column].cat.codes
```

```python
obj_types
```

```python
for column in obj_types:
    df[column] = obj_types[column]
df
```

```python
y = df['Churn']
x = df.drop(['Churn'], 1)
```

```python
#MinMaxScaler scales and translates each feature individually such that 
#it is in the given range on the training set, e.g. between zero and one.

#Used when we have a huge span in difference of data. Makes it easier for the machine to understand
#data and make more precise predictions

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

#logreg is a classification algorithm despite the name.
logreg = LogisticRegression()

logreg.fit(X_train,y_train)

print('Training set score:%.2f'%logreg.score(X_train,y_train))
print('Test set score:%.2f'%logreg.score(X_test,y_test))
```

```python
from sklearn.svm import LinearSVC, SVC

linsvc = LinearSVC()

linsvc.fit(X_train,y_train)

print('Training set score:%.2f'%linsvc.score(X_train,y_train))
print('Test set score:%.2f'%linsvc.score(X_test,y_test))
```

```python
kernel = SVC(kernel = 'rbf', C = 1, gamma = 1)

kernel.fit(X_train, y_train)

kernel_pred = kernel.predict(X_test)

print('Training set score:%.2f'%kernel.score(X_train,y_train))
print('Test set score:%.2f'%kernel.score(X_test,y_test))
```

```python
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier(hidden_layer_sizes = [100,1000],
                  activation = 'relu',
                  solver = 'adam',
                  alpha = 0.001,
                  max_iter = 1000)

nn.fit(X_train,y_train)
print("Train score: ", nn.score(X_train,y_train))

nn.fit(X_test,y_test)
print("Test score: ", nn.score(X_test,y_test))
```
